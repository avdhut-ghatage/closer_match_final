```python
import pandas as pd
import numpy as np
```

```python
company=pd.read_csv('company_scraped.csv')
```

```python
candidates = pd.read_csv('candidate_for_companies.csv')
```

```python
candidates
```

```python
company['profession_match']=company['profession'].str.contains('Doctor|Medical|Practitioner|gp|hospital|dentist|Physician',case=False)
```

```python
company['business_match']=company['nature_of_business'].str.contains('Medical|hospital|health',case=False)
```

```python
company['id']=company.index+1
```

```python
company['Region']=company['Region'].apply(lambda x:str(x).title())
```

```python
company
```

```python
def func(row):
    result=np.nan
    priority=10
    flag= False
    for i in range(len(company['first+last'])):
        
        if row['first+last']==company['first+last'][i] and row['qualification_year']==company['qualification_year'][i]:
            flag= True
            if company['name'][i]!=row['first+last'] and company['name'][i]==row['full_name']:
            
                if row['nhsei_regional_team']==company['Region'][i]:

                    if company['profession_match'][i]==True:
                        return company['id'][i]
                                                          
                    elif company['business_match'][i]==True:
                        result= company['id'][i]
                        priority=1
                                                                    
                    elif priority>3:
                        priority=3
                        result= company['id'][i]  
                                                          
                elif priority>4:
                    result= company['id'][i]
                    priority=5 
            
            elif row['nhsei_regional_team']==company['Region'][i] and priority>2:

                if company['profession_match'][i]==True:
                    result= company['id'][i]
                    priority=2
                                                      
                elif company['business_match'][i]==True:
                    result= company['id'][i]
                    priority=4 
                                                                
                elif priority>6:
                    result= company['id'][i]
                    priority=6
        
        elif flag==True:
            break
    return result
```

```python
candidates['id']=candidates.apply(func,axis=1)
```
